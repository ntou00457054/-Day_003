### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
import xmltodict 

fh = open("./data/64_72hr_CH.xml", "r",encoding='utf-8')#讀xml檔
xml = fh.read()
fh.close()

d = dict(xmltodict.parse(xml))#將xml以dict 的方式表示

location = d['cwbopendata']['dataset']['locations']['location']
for l in location:
    print(l['locationName'])
    
print(len(location)) #ANS:高雄市有38筆溫度資料
'''
d['cwbopendata']['dataset']['locations']['location'][1]['locationName']

需要注意每一層的Data type

他人的發問 : https://www.cupoy.com/qa/club/ai_tw/0000016E62FB84E4000000026375706F795F72656C656173654B5741535354434C5542/00000170842F0C100000002C6375706F795F72656C656173655155455354
'''
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for l in location:
    print('地點 : ' + l['locationName']+'\n')
    time = location[1]['weatherElement'][0]['time'][0]['dataTime']
    print('第一個時間點 : ' + time +'\n') 
    print('第一個溫度 : ' + location[1]['weatherElement'][0]['time'][0]['elementValue']['value']\
          +location[1]['weatherElement'][0]['time'][0]['elementValue']['measures'] +'\n\n')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

location_0_weather = location[0]['weatherElement']

for i in location_0_weather:
    for j in i['time']:
        print('時間點 : ' + j['dataTime'] +'\n')
        print('溫度 : ' + j['elementValue']['value'] + j['elementValue']['measures'] + '\n\n')
    
```
